```python
import requests

url = "https://raw.githubusercontent.com/prateekralhan/Speech2Text-for-Long-Audio-Files/master/conversation.wav"

r = requests.get(url)
with open("sample_speech.wav", "wb") as f:
    f.write(r.content)
```

```python

import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load("sample_speech.wav")

print("Sampling Rate:", sr)
print("Duration:", len(y)/sr)

plt.figure(figsize=(10,4))
librosa.display.waveshow(y, sr=sr)
plt.title("Speech Signal Waveform")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.show()
```
