```python
import requests

url = "https://raw.githubusercontent.com/prateekralhan/Speech2Text-for-Long-Audio-Files/master/conversation.wav"

r = requests.get(url)
with open("sample_speech.wav", "wb") as f:
    f.write(r.content)
```

```python
import librosa
import matplotlib.pyplot as plt

y, sr = librosa.load("sample_speech.wav")

zcr = librosa.feature.zero_crossing_rate(y)
energy = librosa.feature.rms(y=y)

plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.plot(zcr[0])
plt.title("Zero Crossing Rate")

plt.subplot(2,1,2)
plt.plot(energy[0])
plt.title("Short-Time Energy")

plt.tight_layout()
plt.show()
```
