```python
import requests

url = "https://raw.githubusercontent.com/prateekralhan/Speech2Text-for-Long-Audio-Files/master/conversation.wav"

r = requests.get(url)
with open("sample_speech.wav", "wb") as f:
    f.write(r.content)
```

```python
!pip install SpeechRecognition
import speech_recognition as sr

r = sr.Recognizer()

with sr.AudioFile("sample_speech.wav") as source:
    audio = r.record(source)

try:
    text = r.recognize_google(audio)
    print("Recognized Text:", text)
except:
    print("Speech not recognized")
```
